## Упражнение № 23

```python
ser = pd.Series(['Apple', 'Orange', 'Plan', 'Python', 'Money'])

# Функция для проверки количества гласных в слове
def count_vowels(word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    count = 0
    for char in word.lower():
        if char in vowels:
            count += 1
    return count >= 2

filt_ser = ser[ser.apply(count_vowels)]

print(filt_ser)
```

```python
# Загрузка данных
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

mean_min_price = df['Min.Price'].mean()
median_max_price = df['Max.Price'].median()
def replace_missing_min(value):
    if pd.isnull(value):
        return mean_min_price
    else:
        return value
def replace_missing_max(value):
    if pd.isnull(value):
        return median_max_price
    else:
        return value
df['Min.Price'] = df['Min.Price'].apply(replace_missing_min)
df['Max.Price'] = df['Max.Price'].apply(replace_missing_max)
print(df[['Min.Price', 'Max.Price']])
```

## Упражнение № 44

```python
def swap_columns(df, col1, col2):
    colnames = df.columns.tolist()
    ind1, ind2 = colnames.index(col1), colnames.index(col2)
    colnames[ind2], colnames[ind1] = colnames[ind1], colnames[ind2]
    return df[colnames]

df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
df = df[list('cbade')]
df = swap_columns(df, 'c', 'a')
df[sorted(df.columns, reverse=True)]
```

## Упражнение № 59

```python
from scipy.spatial.distance import cdist
df = pd.DataFrame(np.random.randint(1, 100, 40).reshape(10, -1), columns=list('pqrs'), index=list('abcdefghij'))

distances = cdist(df.values, df.values, metric='euclidean')

np.fill_diagonal(distances, np.inf)

nearest_idx = np.argmin(distances, axis=1)

nearest_labels = df.index[nearest_idx]

nearest_distances = np.min(distances, axis=1)

df['nearest_row'] = nearest_labels
df['dist'] = nearest_distances

print(df)
```
